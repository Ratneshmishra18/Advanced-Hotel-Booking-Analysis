```python
#importing the necessary libraries 
import pandas as pd 
import numpy as np

# Load the Dataset 
df = pd.read_csv("/Users/ratneshmishra/Downloads/hotel_bookings 2.csv")

# Drop rows with missing values 
df.dropna(inplace=True)

#Convert data types 
df['reservation_status_date']=pd.to_datetime(df['reservation_status_date'])

# Feature Engineering 
df['stay_duration'] = df['stays_in_weekend_nights'] + df['stays_in_week_nights']
df['is_canceled']=df['is_canceled'].astype(bool)
df['is_repeated_guest']=df['is_repeated_guest'].astype(bool)
```

```python
# Exploratory Data Analysis (EDA)
import matplotlib.pyplot as plt
import seaborn as sns 

# Stay Duratiion Distribution 
plt.figure(figsize=(10, 6))
plt.hist(df['stay_duration'], bins=30,edgecolor = 'k',alpha=0.7)
plt.title(' Stay Duration Distribution')
plt.xlabel('Stay Duration (nights)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Analyze the cancellation rates
cancellation_rate = df['is_canceled'].mean()
print(f"Cancellation Rate : {cancellation_rate:.2%}")




# Visualize Cancellation Rate by Hotel Type
plt.figure(figsize=(10, 6))
sns.barplot(x='hotel', y='is_canceled', data=df, errorbar=None)
plt.title('Cancellation Rate by Hotel Type')
plt.xlabel('Hotel Type')
plt.ylabel('Cancellation Rate')
plt.show()
```

```python
# Analyze booking lead time 
plt.figure(figsize=(10,6))
sns.histplot(df['lead_time'],bins=50,kde=True)
plt.title('Booking Lead Time Distribution')
plt.xlabel('Lead Time (days)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Analyze customer segmentation 
customer_segments = df['customer_type'].value_counts(normalize=True)
print(customer_segments)

# Visualize customer segments
plt.figure(figsize=(10, 6))
sns.countplot(x='customer_type', data=df, palette='Set2')
plt.title('Customer Segmentation')
plt.xlabel('Customer Type')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

```python
# Key Insights 
# Average Stay Duration
average_stay_duration = df['stay_duration'].mean()
print(f"Average Stay Duration : {average_stay_duration:.2f} nights")

```
